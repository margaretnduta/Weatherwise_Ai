## Planting Advisor

```python
def planting_advisor(rain_prediction, rainfall_amount):
    recommendations = []

    if rain_prediction == 1 and rainfall_amount > 10:
        recommendations.append(
            "🌱 Conditions are favorable for planting."
        )
    else:
        recommendations.append(
            "🌱 Wait for sufficient rainfall before planting."
        )

    return recommendations
```

## Irrigation Advisor

```python
def irrigation_advisor(rain_prediction, max_temp):
    recommendations = []

    if rain_prediction == 1:
        recommendations.append(
            "💧 Delay irrigation because rainfall is expected."
        )

    elif max_temp > 32:
        recommendations.append(
            "💧 Increase irrigation due to high temperatures."
        )

    else:
        recommendations.append(
            "💧 Maintain normal irrigation schedule."
        )

    return recommendations
```

## fertilizer advisor

```python
def fertilizer_advisor(rain_prediction):
    recommendations = []

    if rain_prediction == 1:
        recommendations.append(
            "🌿 Delay fertilizer application until after the rain."
        )

    else:
        recommendations.append(
            "🌿 Suitable conditions for fertilizer application."
        )

    return recommendations
```

## Spraying Advisor

```python
def spraying_advisor(wind_speed):
    recommendations = []

    if wind_speed > 35:
        recommendations.append(
            "🚜 Avoid spraying pesticides due to strong winds."
        )

    else:
        recommendations.append(
            "🚜 Suitable conditions for spraying."
        )

    return recommendations
```

## Pest and disease Advisor

```python
def disease_risk_advisor(humidity, temperature):
    recommendations = []

    if humidity > 85 and 20 <= temperature <= 30:
        recommendations.append(
            "🦠 High fungal disease risk. Monitor crops closely."
        )

    elif humidity < 40 and temperature > 32:
        recommendations.append(
            "🐛 High pest risk due to hot and dry conditions."
        )

    else:
        recommendations.append(
            "✅ Low pest and disease risk."
        )

    return recommendations
```

## Crop Advisor

```python
def crop_advisor(crop_name):
    return [
        f"🌽 Recommended crop: {crop_name.capitalize()}."
    ]
```

## Smart alerts

```python
def smart_alerts(rain_prediction, max_temp, wind_speed):
    alerts = []

    if rain_prediction == 1:
        alerts.append("⚠ Heavy rain expected.")

    if max_temp > 35:
        alerts.append("⚠ Heat stress warning.")

    if wind_speed > 35:
        alerts.append("⚠ Strong winds expected.")

    return alerts
```

```python
def generate_recommendations(
    crop_name,
    rain_prediction,
    rainfall_amount,
    temperature,
    humidity,
    wind_speed
):

    recommendations = []

    recommendations.extend(
        crop_advisor(crop_name)
    )

    recommendations.extend(
        planting_advisor(
            rain_prediction,
            rainfall_amount
        )
    )

    recommendations.extend(
        irrigation_advisor(
            rain_prediction,
            temperature
        )
    )

    recommendations.extend(
        fertilizer_advisor(
            rain_prediction
        )
    )

    recommendations.extend(
        spraying_advisor(
            wind_speed
        )
    )

    recommendations.extend(
        disease_risk_advisor(
            humidity,
            temperature
        )
    )

    recommendations.extend(
        smart_alerts(
            rain_prediction,
            temperature,
            wind_speed
        )
    )

    return recommendations
```

```python
result = generate_recommendations(
    crop_name="maize",
    rain_prediction=1,
    rainfall_amount=20,
    temperature=27,
    humidity=88,
    wind_speed=18
)

for recommendation in result:
    print(recommendation)
```
